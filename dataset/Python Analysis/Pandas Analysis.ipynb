# handle missing values

```python
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Not Rated")
# change the  datatype of the date from string to datetime
df['date_added']=pd.to_datetime(df['date_added'])
```

```python
df.dtypes
```

```python
df['year_added']=df['date_added'].dt.year
```

# movies vs TV shows

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure()
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows Count")
plt.show()
```

```python
plt.figure()
plt.figure(figsize=(12,6))
sns.countplot(x='rating' , data=df , order=df['rating'].value_counts().index)

plt.title("Rating Distribution")
plt.show()
```

# Top 10 countries 

```python
top_countries = df['country'].value_counts().head(10)

plt.figure()
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries")
plt.show()
```

# Top 10 generes

```python
genres = df['listed_in'].str.split(',', expand=True).stack().str.strip()
top_genres = genres.value_counts().head(10)

plt.figure()
sns.barplot(x=top_genres.values, y=top_genres.index)
plt.title("Top 10 Genres")
plt.show()
```

# Content added per year

```python
df_temp = df.dropna(subset=['date_added']).copy()
df_temp['year_added'] = df_temp['date_added'].dt.year

plt.figure()
df_temp['year_added'].value_counts().sort_index().plot()
plt.title("Content Added Per Year")
plt.show()
```

# movies and Tv show added per year

```python
# null  date remove from data
df_temp=df.dropna(subset=['date_added']).copy()
df_temp['year_added'] = df_temp['date_added'].dt.year
plt.figure()
sns.countplot(x='year_added', hue='type', data=df_temp)

plt.xticks(rotation=90)
plt.title("Movies vs TV Shows Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

# Movie Duration Distribution

```python
movies = df[df['type'] == 'Movie'].copy()
movies['duration'] = movies['duration'].str.replace(' min','').astype(float)

plt.figure()
sns.histplot(movies['duration'], bins=30)
plt.title("Movie Duration Distribution")

plt.show()
```

# tv show season distribution 

```python
tv = df[df['type'] == 'TV Show'].copy()
tv['duration'] = tv['duration'].str.replace(' Seasons','').str.replace(' Season','').astype(float)

plt.figure()
sns.histplot(tv['duration'], bins=10)
plt.title("TV Show Seasons Distribution")
plt.show()
```

# Different Rating rate to movies and tv shows

```python
plt.figure()
sns.countplot(y='rating', hue='type', data=df)
plt.title("Rating vs Type")
plt.show()
```
